```python
import sys
import os
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
from selenium import webdriver
import time
import datetime as dt
import tqdm
from tqdm.notebook import tqdm

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
#크롬브라우저 실행
path= 'C:\\Users\\qkd78\\Desktop\\방성현\\dwellon\\소비문화 크롤링\\chromedriver.exe'
browser= webdriver.Chrome(path)
```

```python

timespace = 1
startdate=dt.date(year=2021, month=8, day=1)
enddate=dt.date(year=2019, month=8, day=1)
untildate=startdate-dt.timedelta(days=timespace)
```

```python
lastpgnum=100 #마지막 페이지 넘버 기입하기
cafe_dict= {'date': [], 'cafe': [], 'title': [], 'content': []}

check=""
keyword = ['전기차']
for keyword in keyword:
    
    while not enddate >= untildate:
        

        pgnum=1
        character = "-"

        edit_startdate = str(startdate)
        for x in range(len(character)):
            edit_startdate = edit_startdate.replace(character[x], ".")

        edit_enddate = str(enddate)
        for x in range(len(character)):
            edit_enddate = edit_enddate.replace(character[x], ".")

        edit_untildate = str(untildate)
        for x in range(len(character)):
            edit_untildate = edit_untildate.replace(character[x], ".")



        while pgnum < lastpgnum+1:

            print('{0}~{1}'.format(untildate, startdate))
            url = "https://cafe.naver.com/ca-fe/home/search/articles?q="+keyword+"&em=1&p="+str(pgnum)+"&pr=7&ps="+edit_untildate+"&pe="+edit_startdate
            browser.get(url)
            browser.implicitly_wait(5) 
            try:
                WebDriverWait(browser, 5).until(EC.presence_of_element_located((By. CLASS_NAME, 'item_subject'))
                )

            except:
                pass


            html=browser.page_source
            soup=BeautifulSoup(html,'html.parser')


            #browser.implicitly_wait(5) 
            #print(html)
            try:
                checkif=browser.find_element_by_class_name('item_subject').text

            except:
                break

            if checkif==check:
                print("다음 기간으로 넘어갑니다")
                break

            date=soup.find_all("span",{"class": "write_date_time"})
            for k in date:
                cafe_dict['date'].append(k.text)

            cafe=soup.find_all("span",{"class": "cafe_name"})
            for k in cafe:
                cafe_dict['cafe'].append(k.text)

            title=soup.find_all("a",{"class": "item_subject"})
            for k in title:
                cafe_dict['title'].append(k.text)

            content=soup.find_all("p",{"class": "item_content"})
            for k in content:
                cafe_dict['content'].append(k.text)


            #length= len(cafe_dict['date'])

            #print(f'--------{pgnum}페이지 결과입니다.---------')
           # print(len(cafe_dict['title']))

            #print(data_list)
            print()
            #print(len(data_list))

            pgnum+=1

            check= browser.find_element_by_class_name('item_subject').text
            


        startdate=untildate
        untildate=startdate-dt.timedelta(days=timespace)
    
    a=list(cafe_dict['title'])
    b=list(cafe_dict['date'])
    c=list(cafe_dict['cafe'])
    d=list(cafe_dict['content'])
    f=len(a)-len(b)
    j = len(a)-len(c)
    k = len(a)-len(d)
    for x in range(f):
        cafe_dict['date'].append('삭제요망')
    for x in range(j):
        cafe_dict['cafe'].append('삭제요망')
    for x in range(k):
        cafe_dict['content'].append('삭제요망')
    df = pd.DataFrame(cafe_dict)
    print(df.head())
    df.to_csv('%s 네이버 카페.csv'%keyword)
```
